### 2.1- Razonamiento
Desplegar un modelo razonador y parametrizar distintos grados de razonamiento (low, medium, high)

```python
from openai import OpenAI

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT_p2")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT_p2")
api_key = os.getenv("AZURE_OPENAI_API_KEY_p2")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "Give me the instructions to start to invest in the stock market",
        }
    ],
    max_completion_tokens=5000,
    reasoning_effort="low",  
)

print(completion.choices[0].message.content)
```

### 2.2- Function calling
Activar un motor de búsqueda web para probar llamadas a funciones (`function calling`) que recuperen información externa.

```python
from openai import OpenAI

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT_p2")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT_p2")
api_key = os.getenv("AZURE_OPENAI_API_KEY_p2")

client = OpenAI(
    api_key=api_key,
    base_url=endpoint,
)

response = client.responses.create(   
  model=deployment_name,
  tools=[{"type": "web_search"}], 
  input="Please perform a web search on the results of the 30th matchday of LaLiga 2025–26"
)

print(response.output_text)
```
