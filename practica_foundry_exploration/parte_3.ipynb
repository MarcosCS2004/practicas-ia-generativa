## 3) Modelos Multimodales
### Objetivo: 
Desplegar un modelo multimodal y probar interacciones que involucren imágenes, audio y/o texto combinado (por ejemplo: describir una imagen, transcribir audio y responder preguntas sobre su contenido, etc.).

Imagen enviada


```python
from openai import OpenAI

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT_p1")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT_p1")
api_key = os.getenv("AZURE_OPENAI_API_KEY_p1")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

image_url = "https://preview.redd.it/my-absolute-favorite-depiction-of-spinosaurus-river-dragon-v0-ecpk1blgvsab1.jpg?width=640&crop=smart&auto=webp&s=a6dc9d50d001cf21d65693902f67d6f9ac4b56e9"

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {"role": "system", "content": "You are a helpful assistant that can generate responses based on images."},
        {"role": "user", "content": [
            {"type": "text", "text": "Can you describe the image that I send you?"},
            {"type": "image_url", "image_url": {"url": image_url}}
        ]}
    ],
    temperature=1,
    max_tokens=2048,
)

print(completion.choices[0].message.content)
```
