### Rol
**EXPLICACION:** Darle a la IA una instrucción para que tome un papel en especififco.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "\"You are a history professor talking to a student.\" explain to me the ancient history",
        }
    ],
)
print(completion.choices[0].message.content)
```

**ANALISIS:** La respuesta ha sido existosa dando un pequeño resumen del tema dado.Esta tecnica es util para ecplicaciones o roles de asistencia.

### Few-shot
**EXPLICACION:** Dar ejemplos de input y output para controlar el patron de respuesta

```python
import os
from dotenv import load_dotenv
#EJEMPLO
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "Tell me the points that each team gets with the match results"
            "Examples:"
            "Team A 1-0 Team B->Team A: 3 points|Team B: 0"
            "Team A 1-1 Team B->Team A: 1 points|Team B: 1"
            "Team A 0-1 Team B->Team A: 0 points|Team B: 3"
            "Now give me the points of this result Team A 3-2 Team B",
        }
    ],
)
print(completion.choices[0].message.content)
```

**ANALISIS:** La respuesta ha sido correcta un poco repetitiva.Esta forma resulta muy util para analisis de textos y clasificacion de resultados.

### Zero-shot
**EXPLICACION:** Dar una orden simple sin ejemplos

```python
import os
from dotenv import load_dotenv
#EJEMPLO
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "Tell me the points that each team gets with the match results knowing win 3 points,draw 1,lose 0"
            "Give me the points of this result Team A 5-1 Team B",
        }
    ],
)
print(completion.choices[0].message.content)
```

### Cues
**EXPLICACION:** Poner el inicio de la respuesta que se espera.

```python
import os
from dotenv import load_dotenv
#EJEMPLO
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "Summarize the following text in 3 words."
            "Lucas goes to school every day of the week. He has many subjects to go to each school day: English, art, science, mathematics, gym, and history. His mother packs a big backpack full of books and lunch for Lucas."
            "Answer:"
            "Main point:"
            "Second Point:"
            "Third Point:"
        }
    ],
)
print(completion.choices[0].message.content)
```

**ANALISIS:** La respuesta ha sido medio correcta no ha puesto los "point:" antes de la respuesta pero si ha dado las ideas principales.Le veo utilidad en lectura de contratos,mails o textos extensos para sacar las ideas importantes

### Chain-of-thought
**EXPLICACION:** Poner pasos a paso como se quiere la respuesta

```python
import os
from dotenv import load_dotenv
#EJEMPLO
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "If a train travels 60 miles in 1 hour, how far will it travel in 3 hours?"
            "Let's think step by step."
        }
    ],
)
print(completion.choices[0].message.content)
```

### Ejemplo 1 — SOLO Rol
**Caso de uso: Generación de un resumen ejecutivo profesional**

La técnica de Rol es ideal cuando necesitas que el modelo adopte un estilo, tono o comportamiento específico.
En este caso, queremos un resumen ejecutivo profesional, por lo que asignar el rol de Consultor de Negocios Senior garantiza los minimos necesarios para este tipo de documentos.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": """
Actúa como un Consultor de Negocios Senior especializado en análisis estratégico.
Tu tarea es elaborar un resumen ejecutivo claro, conciso y orientado a la toma de decisiones.

Resume el siguiente texto:
"La empresa ha experimentado un crecimiento del 12% anual, pero enfrenta problemas de rotación de personal y retrasos en la cadena de suministro."
"""
        }
    ],
)

print(completion.choices[0].message.content)
```

### Ejemplo 2 — SOLO Few‑Shot
**Caso de uso: Clasificación de sentimiento**
La técnica Few‑Shot es ideal cuando:
quieres que el modelo siga un formato específico
necesitas consistencia en la clasificación
deseas reducir ambigüedad mostrando ejemplos claros


```python
import os
from dotenv import load_dotenv
#EJEMPLO
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": """
Actúa como un Analista de NLP experto en clasificación de sentimiento. 
Tu tarea es identificar el sentimiento y justificarlo brevemente.

Ejemplo 1:
"El producto llegó rápido y funciona perfecto."
Sentimiento: Positivo
Justificación: Expresa satisfacción con la calidad y la entrega.

Ejemplo 2:
"No es lo que esperaba, pero tampoco está mal."
Sentimiento: Neutro
Justificación: Mezcla de aspectos positivos y negativos.

Ejemplo 3:
"La batería dura muy poco y el servicio técnico no responde."
Sentimiento: Negativo
Justificación: Menciona problemas y mala experiencia.

Clasifica el sentimiento de la siguiente reseña:
"El diseño es bonito, pero el rendimiento es terrible."
"""
        }
    ],
)

print(completion.choices[0].message.content)
```
