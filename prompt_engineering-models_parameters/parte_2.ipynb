### 2.1 - Experimentación con temperature
Prueba el **mismo prompt** con diferentes valores de `temperature`:
- `temperature = 0.0`
- `temperature = 0.5`
- `temperature = 1.0`
- `temperature = 1.5` (si la API lo permite)

Analiza cómo cambian las respuestas. Usa un caso práctico (ejemplo: generar un eslogan, escribir código, responder una pregunta técnica).

La temperatura controla qué tan creativas o predecibles son las respuestas de un modelo de IA.
Con valores bajos, el modelo responde de forma más segura y consistente; con valores altos, genera respuestas más variadas y arriesgadas.
En esencia, ajusta cuánto explora el modelo opciones menos probables al generar texto.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables from .env file
load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT")
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "Cuentame una pequeña historia en una linea",
        }
    ],
    temperature = 0.0
)

print(completion.choices[0].message.content)
```

top_p es otro parámetro que controla la creatividad, pero funciona distinto a la temperatura.

Incluye una sección en el notebook respondiendo estas preguntas basándote en tu experiencia práctica y lo aprendido:

1. **¿Cuál es la diferencia entre top_p y temperature?**

Temperature ajusta la creatividad general, mientras que top_p controla qué parte del núcleo de palabras probables se considera.

2. **¿Por qué no se recomienda ajustar top_p y temperature al mismo tiempo?**

Porque ambos modifican la aleatoriedad, pero desde ángulos distintos, y combinarlos puede producir comportamientos impredecibles o difíciles de controlar.
Si los dos están activos, es más complicado entender cuál de los dos está afectando el resultado y ajustar el modelo de forma estable.
Por eso se recomienda usar uno u otro, no ambos a la vez.

3. **¿Cuál es la diferencia entre presence_penalty y frequency_penalty?**

Presence_penalty evita repetir temas; frequency_penalty evita repetir palabras en exceso.
